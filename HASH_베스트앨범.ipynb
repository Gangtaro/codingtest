```python
def solution(genres, plays):
    answer = []
    
    dicSum = dict()
    dicMapping = dict()
    Info = []

    for i,genre in enumerate(genres):
        if dicSum.get(genre) == None:
            dicSum[genre] = plays[i]
            dicMapping[genre] = len(Info)
            Info.append([])
            Info[dicMapping[genre]].append([plays[i],-i])
        else:
            dicSum[genre] += plays[i]
            Info[dicMapping[genre]].append([plays[i],-i])


    dicSum = sorted(dicSum.items(), key= lambda dicSum: dicSum[1], reverse=True)
    for k in Info: 
        k.sort(reverse=True)
    
    for value in dicSum:
        for i,items in enumerate(Info[dicMapping[value[0]]]):
            if i >1 : break
            answer.append(items[1]*-1)
    return answer
```
