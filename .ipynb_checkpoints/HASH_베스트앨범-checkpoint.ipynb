```python
import numpy as np
import pandas as pd

def solution(genres, plays):
    musics = pd.DataFrame()
    musics['id'] = range(len(plays))
    musics['plays'] = plays
    musics['genres'] = genres
    
    musics = pd.merge(musics, musics.groupby('genres').sum().reset_index().drop('id',axis =1), on = 'genres')

    order_genre = musics.groupby('genres').head(1)[['genres','id']].sort_values('id')
    order_genre['order_genre'] = range(len(order_genre))
    order_genre = order_genre.drop('id', axis =1)

    musics = pd.merge(musics, order_genre, on ='genres')
    musics['order_genre'] = musics['plays_y']*100-musics['order_genre']
    musics.sort_values('plays_x', ascending =False).sort_values('order_genre', ascending =False)
    answer = musics.sort_values('plays_x', ascending =False).sort_values('order_genre', ascending =False).groupby('genres').head(2)['id']
    answer = list(answer)
    answer
   
    return answer
```

```python
import numpy as np
import pandas as pd


musics = pd.DataFrame()
musics['id'] = range(len(examples[4]['genres']))
musics['plays'] = examples[4]['plays']
musics['genres'] = examples[4]['genres']

# 장르 별 재생횟수 합에 관한 변수 생성: genres_plays
musics['genres_plays'] = musics.groupby('genres').transform(np.sum)['plays'] 

# 장르 별 고유번호가 빠른 순으로 장르 순번에 관한 변수 생성: genres_head_id
musics['genres_head_id'] = musics.groupby('genres').transform(np.min)['id']

# 장르 별 재생횟수 -> (빠른 고유번호를 가진 장르 순번대로 재배치) -> 각 음원의 재생횟수 순으로 데이터프레임 순서를 매기는 변수 생성
musics['order'] = musics['genres_plays']*1000000 - musics['genres_head_id']*1000 + musics['plays'] 

# 각 장르 별로 2개의 음악의 고유변수 추출
answer = musics.sort_values('order',ascending=False).groupby('genres').head(2)['order']
answer = list(answer)

answer
```

```python
def solution1(genres, plays):
    answer = []
    
    dicSum = dict()
    dicMapping = dict()
    Info = []

    for i,genre in enumerate(genres):
        if dicSum.get(genre) == None:
            dicSum[genre] = plays[i]
            dicMapping[genre] = len(Info)
            Info.append([])
            Info[dicMapping[genre]].append([plays[i],-i])
        else:
            dicSum[genre] += plays[i]
            Info[dicMapping[genre]].append([plays[i],-i])


    dicSum = sorted(dicSum.items(), key= lambda dicSum: dicSum[1], reverse=True)
    for k in Info: 
        k.sort(reverse=True)
    
    for value in dicSum:
        for i,items in enumerate(Info[dicMapping[value[0]]]):
            if i >1 : break
            answer.append(items[1]*-1)
    return answer
```

- list:genre => key, list:plays => value 로 구성되는 dict를 생성  
    - zip() 함수를 통해서 dict 만들 수 있다. [출처](https://security-nanglam.tistory.com/427)
- 중복된 key를 가진 dict를 생성할 수 없다
- 문제상황과는 다르게 재생횟수가 일치하는 장르에 대해서도 고려해줘야 한다. 
해결: 상황에 맞는 모든 것을 포함한 선형식 통해 정렬하는 방법을 채택했다.  
    => 우선순위대로 
        1. 재생횟수가 가장 많은 장르 순으로 정렬
        2. 같은 재생횟수를 공유하는 장르는 보유한 고유번호를 비교했을때 더 빠른 고유번호를 보유하는 장르를 우선 정렬
        3. 같은 장르 내에서는 재생 횟수가 높은 순으로 정렬  
    이를 해결하기 위해서 모든 정보를 잃지 않을만큼의 순서대로 큰 값을 곱한 선형식을 통해서 하나의 값을 만든다.  
    이 값을 정렬하면 모든 정보를 지닌채 정렬가능하다.  
    그 다음 각 장르별로 최대2개를 뽑는 절차를 진행하면 결과도출이 가능하다.

#### 효율적인 풀이를 통해서 해시 학습

- **sorted** [참고](http://blog.weirdx.io/post/50236)
    - iterable data 에 대해서 동작한다

        2. key 파라미터로 전달된 함수는 입력 레코드마다 한번씩 호출되어진다.
        3. 일반적 패턴은 클래스나 복잡한 객체들을 정렬할 때 사용한다.

        + **Operator Module Functions**

               - itemgetter()

```python
# sorted([iterable],key = function)

fruit = ['apple', 'banana', 'pear', 'orange', 'corn']
count = [30, 70, 80, 20, 10]
fruits = dict()

for i in range(len(fruit)) :
    fruits[fruit[i]] = count[i]

# fruits
print('fruits: ',fruits)

# sorted
sorted(fruits)
print('sorted(fruits): ',sorted(fruits))

# sorted( ,key = str.lower)
print('sorted(fruits): ',sorted(fruits, key = str.lower))

# sorted( ,key = str.lower)
print('sorted(fruits): ',sorted(fruits, key = str.lower))

# operator
from operator import itemgetter, attrgetter
print(sorted(fruits, key = itemgetter(1)))

    
```
